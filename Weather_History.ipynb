```python
# Get the hisotrical weanther data for games played
import pandas as pd
import requests
import time
from pprint import pprint
from mike_api import api_key
# create code to test the WorldWeatherOnline API


# url = HTTPS: https://api.worldweatheronline.com/premium/v1/past-weather.ashx
# example request url . 
#URL http://api.worldweatheronline.com/premium/v1/past_weather.ashx?q=London&amp;format=xml&amp;key=xxxxxxxxxxxxxxxxx
game_data_toload = "stadiums_merge_type.csv"
game_data = pd.read_csv(game_data_toload)
game_data.head()


cityid = game_data['CITY']
game_date = game_data['SCHEDULE DATE']
#game_date = ['2014-09-07', '2014-09-14', '2014-09-21', '2014-09-28']
temp = []
description = []
weather_code = []
location = []
date = []
game_date
#start = time.mktime(datetime.datetime.strptime(game_date, "%d/%m/%Y").timetuple())

# format_url = f"{url}{cityid}&date={game_date}&format=json&key={api_key}"
#print(format_url)
```

```python
url = 'https://api.worldweatheronline.com/premium/v1/past-weather.ashx?q='

count = 0
#request_url = f"{url}{cityid}&date={game_date}&format=json&key={api_key}"
#weather = requests.get(request_url).json()

# loop in loop to request weather data for each NFL city on each game day.

for day in game_date:
    try:
        for city in cityid:
            try:
                count += 1
                request_url = f"{url}{city},US&date={day}&format=json&key={api_key}"
                weather = requests.get(request_url).json()
                location.append(city)
                date.append(day)
                temp.append(weather['data']['weather'][0]['hourly'][6]['FeelsLikeF'])
                description.append(weather['data']['weather'][0]['hourly'][6]['weatherDesc'][0]['value'])
                weather_code.append(weather['data']['weather'][0]['hourly'][6]['weatherCode'])
                #pprint(weather)
            except:
                print("No game in City this week")
    except:
        print("No date found")
print(temp)
print(description)
print(weather_code)
```

```python
# Create a dictionary from the weather data lists. Used to create dataframe


weather_dict = {
    'Game Date': date,
    'City': location,
    'Temperature': temp,
    'Weather': description, 
    'Code': weather_code
}
#weather_dict

# create a data frame from arrays filled by responses
weather_df = pd.DataFrame(weather_dict)
weather_df.to_csv('weather_data.csv')
weather_df
```
